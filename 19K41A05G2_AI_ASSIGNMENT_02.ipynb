```python
import numpy as np
x=np.random.uniform(-10,10)
eta=0.001
iter_max=300
```

```python
 iter=1
 while iter<=iter_max :
  slope = 4*x*x*x + 6*x
  step_length = (-eta) * slope
  x=x+step_length
  iter=iter+1 
 print("control variable = ",round(x))
 obj = ((x*x*x*x)+(3*x*x)+10)
 print("objective function = ",round(obj))
```
